```python
valor_compra = int(input("Digite o valor total da sua compra: R$ "))

if valor_compra < 50:
    print("Obrigado por comprar conosco!")
elif 50 <= valor_compra <= 99:
    print("Parabéns! Você ganhou um brinde!")
elif 100 <= valor_compra <= 199:
    print(f"Desconto de 10 reais aplicado!\nValor a pagar: R$ {valor_compra - 10}")
else:
    print(f"Desconto de 25 reais aplicado!\nValor a pagar: R$ {valor_compra - 25}")
```

```python
entrada = input()
valor_transacao, taxa_servico, pagamento_minimo = map(int, entrada.split())

valor_final = valor_transacao - taxa_servico

if valor_final >= pagamento_minimo:
    print("Aprovada")
else:
    print("Recusada")
```

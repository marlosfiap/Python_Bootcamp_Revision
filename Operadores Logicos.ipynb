```python
# tabela verdade com operadores E e OU e suas NEGAÇÕES
tabela_verdade = pd.DataFrame({
    "A": [False, False, True, True],
    "B": [False, True, False, True],
    "A and B": [False, False, False, True],
    "A or B": [False, True, True, True],
    "not A": [True, True, False, False],
    "not B": [True, False, True, False],
})

tabela_verdade
```

```python
# Valores para utilização em exemplos
saldo = 1000
saque = 250
limite = 200
conta_especial = True
conta_normal = False
```

```python
# Na tabela verdade do operador E (AND) - para ser verdadeiro,
# as duas sentenças precisam necessariamente ser verdadeiras.

saldo >= saque and limite <= saque
```

```python
# Na tabela verdade do operador OU (OR) - para ser verdadeiro
# qualquer operador precisa ter algum valor verdadeiro. Logo,
# para ser falso, as duas sentenças precisam ser falsas.

saque <= limite or conta_normal
```

```python
# Para entender operadores lógicos a leitura é realizada da esquerda para
# direita, mas, existe a preferÊncia em caso do uso de parentesis.

saldo >= saque and saque <= limite or conta_especial and saldo >= saque
```

```python
(saldo >= saque and saque <= limite) or (conta_especial and saldo >= saque)
```
